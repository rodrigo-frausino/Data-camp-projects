```python
# So I can display multiple outputs in a single cell
from IPython.core.interactiveshell import InteractiveShell

InteractiveShell.ast_node_interactivity = "all"
```

```python
# Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt
# Load the CSV file and store as netflix_df
netflix_df=pd.read_csv("netflix_data.csv")
netflix_df.head()
```

```python
#Filter the data to remove TV shows and store as netflix_subset
print(netflix_df.type.value_counts()) # just to see how many types we have.
netflix_subset=netflix_df[netflix_df["type"]=="Movie"]
netflix_subset.head()
```

```python
# Investigate the Netflix movie data, keeping only the columns "title", "country", "genre", "release_year", "duration", and saving this into a new DataFrame called netflix_movies
netflix_movies=netflix_subset[['title','country','genre','release_year','duration']]

netflix_movies.head()
```

```python
#Filter netflix_movies to find the movies that are strictly shorter than 60 minutes, saving the resulting DataFrame as short_movies; inspect the result to find possible contributing factors.
short_movies=netflix_movies[netflix_movies["duration"]<60]
short_movies['genre'].value_counts()
short_movies.head()
```

```python
#Using a for loop and if/elif statements, iterate through the rows of netflix_movies and assign colors of your choice to four genre groups 
#("Children", "Documentaries", "Stand-Up", and "Other" for everything else). Save the results in a colors list. Initialize a matplotlib figure 
#object called fig and create a scatter plot for movie duration by release year using the colors list to color the points and using the labels 
#"Release year" for the x-axis, "Duration (min)" for the y-axis, and the title "Movie Duration by Year of Release"

colors=[]
for lab, row in netflix_movies.iterrows():
    if row['genre']=='Children':
        colors.append('Red')
    elif row['genre']=='Documentaries':
        colors.append('Green')
    elif row['genre']=='Stand-Up':
        colors.append('Blue')
    else:
        colors.append('Black')

fig=plt.figure(figsize=(12,8))

plt.scatter(netflix_movies.release_year,netflix_movies.duration,c=colors)
plt.title('Movie Duration by Year of Release')
plt.xlabel('Release year')
plt.ylabel('Duration (min)')
plt.show()

```

```python
movies_release_duration=netflix_movies[netflix_movies['duration']>60] #Excluding short movies
movies_release_duration=movies_release_duration[['release_year','duration']]

movies_release_duration=movies_release_duration.groupby('release_year').agg(['mean','std'])
movies_release_duration=movies_release_duration.reset_index()

plt.errorbar(movies_release_duration['release_year'], movies_release_duration[('duration','mean')], yerr=movies_release_duration[('duration','std')],fmt='o',markersize=4, capsize=3,color='black')
plt.ylabel('Average duration of movie (min)')
plt.xlabel('Year of Release')
```
